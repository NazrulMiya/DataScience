# Description  

**Support-vector machines (SVMs, also support-vector networks)** are supervised learning models is used for classification and regression analysis.  
Given a set of training examples, each marked as belonging to one or the other of two categories, an SVM training algorithm builds a model that assigns new examples to one category or the other.  

It is **non probabilistic classifier** because it does not use probability distribution over classes to predict, instead it predicts most likely class.    

In the case of support-vector machines, a data point is viewed as a **p-dimensional vector** (a list of p features) known as **'Support Vectors'** . SVM tries to find a hyperplane which can classify datapoints. 
<img src='SVM_1.png' style="height: 200px">  
There are many hyperplanes that might classify the data. Look at the figure above , hyperplannes $H_1$, $H_2$ and $H_3$ can classify the data. then which hyperplane to be selected as best hyperplane for classifying?  
One reasonable choice as the best hyperplane is the one that represents the largest separation, or margin, between the two classes.SVM choose the hyperplane so that the distance from it to the nearest data point on each side is maximized. This is known as **maximum-margin hyperplane**. In the figure $H_3$ is the maximum-margin hyperplane.  **Margin:** Given a particular hyperplane, we can compute the distance between the hyperplane and the closest data point. Once we have this value, if we double it we will get what is called the margin.

If an hyperplane is very close to a data point, its margin will be small. The further an hyperplane is from a data point, the larger its margin will be. The optimal hyperplane will be the one with the biggest margin. The objective of the SVM is to find  the optimal separating hyperplane which maximizes the margin of the training data.the objective of the SVM is to find  the optimal separating hyperplane which maximizes the margin of the training data.  

> ### Hyperplane  
A line equation is $Y = aX + b$ which can be written as $Y - aX - b = 0$, same thing can be represented as vector notation $\vec{W}^T\vec {X} = 0 $ where $$ \vec {W} = \begin{bmatrix} -b \\ -1 \\ 1 \end{bmatrix}  ,  \vec {X} = \begin{bmatrix} 1 \\ x \\ y \end{bmatrix}$$ $\vec{W}$ is Weight Vector ,holds the coeficients of each feature. This weight vector always to be the normal to hyperplane. when SVM defines hyperplane , it supposes a weight vector that is Orthogonal to the hyperplane.  

> ### Orthogonal Projection of a Vector  
Given two vectors $\vec{x}$ and $\vec{y}$, we would like to find the orthogonal projection of $\vec{x}$ onto $\vec{y}$. When we project $\vec{x}$ onto $\vec{y}$ we get another vector $\vec {z}$   
By definition , $Cos\theta = \frac{|z|}{|x|}$  
=> $|z| = |x|Cos\theta$  
=>$|z| = |x|. \frac {x.y}{|x||y|}$  
=>$|z| = \frac {x.y}{|y|}$  
If we define the vector $\vec{u}$ as the direction of $\vec{y}$ then $u = \frac{y}{|y|}$. Hence,  
=>$|z| = x.u$  
Now $|z|$ is the norm of Z vector. Again $\vec{u}$  is in the directon of $\vec{y}$ as in the direction of $\vec{z}$ . so we can say $u = \frac{z}{|z|}$ => $|z| = \frac{z}{u}$  
Hence, =>$|z| = x.u$  
=> $\frac{z}{u} = x.u$  
=>$z = (x.u)u$  
**Finally we can say $\vec{z}=(\vec{u} . \vec{x})\vec{u}$ is the orthogonal projection of $\vec{x}$ onto $\vec{y}$.**  
**Orthogonal Projection helps us to find distance between data point and hyperplane**  

#### Margin Calculation
SVM uses the Orthogonal  Projection of datapoint onto weight vector to find the distance between the datapoint and hyperplane.Lets understand this with an example,  
in the figure below SVM finds hyperplane with equation $X_2 = -2 X_1$ which seperates two groups of data.

Algorithm has to then find the closest data point to the hyperplane. so SVM starts with data point calculating distance between data point A (3,4) and hyperplane. From the hyperplane equation , weight vector $$\vec {W} = \begin{bmatrix} 2 \\ 1 \end{bmatrix} \vec {X} = \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$$.  
We can view the point A as a vector from the origin to A. If we project it onto the normal vector $\vec{w}$ We get the vector $\vec{p}$. Our goal is to find the distance between the point A(3,4) and the hyperplane.
<img src='SVM_4.png' style="height: 300px">
We can see in figure that this distance is the same thing as $|p|$. Lets compute $|p|$.  
We start with two vectors, $\vec{w}=(2,1)$ which is normal to the hyperplane, and $\vec{a} = (3,4)$ which is the vector between the origin and A, then $|w|= \sqrt{2^2+1^2}$  

Let the vector $\vec{u}$ be the direction of $\vec{w}$ so $u = (\frac{2}{\sqrt{5}}, \frac{1}{\sqrt{5}})$  

$\vec{p}$ is the orthogonal projection of $\vec{a}$ onto $\vec{w}$ so :  
$p = (a.u)u$  
after replacing $\vec{u}=(\frac{2}{\sqrt{5}}, \frac{1}{\sqrt{5}})$ and $\vec{a}=(3,4)$ in this equation we get $\vec{p} = (4,2)$  
Hence $|p| = \sqrt{4^2+2^2} = 2\sqrt{5}$, **Hence the distance between A(3,4) and hyperplane is $2\sqrt{5}$**.

**Margin** = $2|p|=2.2\sqrt{5}=4\sqrt{5}$.  

#### Optimal Hyperplane
We have seen how to calculate Margin on either side of hyperplane. $Margin = 2|P|$. Now the question is how to find optimal hyperplane from many possible hyperplanes?.   

We are given a training dataset of $n$ points of the form $(\vec{x_1}, y_1), (\vec{x_2}, y_2),......,(\vec{x_n}, y_n)$. where $y_i$ is the are either 1(Positive Class) or −1(Negetive Class), each indicating the class to which the point $\vec {x_i}$ belongs. Each $\vec {x_i}$ is a $p-dimensional$ real vector. We want to find the "maximum-margin hyperplane or optimal hyperplane" that divides the group of points $\vec {x_i}$ for which $\vec {y_i} = 1$ from the group of points for which $\vec {y_i} = -1$,   

A hyperplane equation in a $2$-Dimensional Vector space is written as $\vec{W}.\vec{X} - b = 0$ where $\vec{W}$ is the normal vector to the hyperplane.  

##### Hard Margin
If the training data is linearly separable, SVM selects two parallel hyperplanes that separate the two classes of data, so that the distance between them is as large as possible. The region bounded by these two hyperplanes is called the "margin".These two hyperplanes can be described by the equations,  

$\vec{W}.\vec{X} - b = 1$ (For group  of points belong to positive class)  
$\vec{W}.\vec{X} - b = -1$ (For group  of points belong to negetive class)  


<img src='SVM_6.png' style="height: 300px">   

We also have to prevent data points from falling into the margin, we add the following constraint: for each $i$ either,  

$\vec{W}.\vec{X} - b \geq 1$, if $y_i = 1$ ,  
or
$\vec{W}.\vec{X} - b \leq 1$, if $y_i = -1$ ,

These constraints state that each data point must lie on the correct side of the margin.  
these two constraints combiningly can be written as  

$y_i (\vec{W}.\vec{X} - b) \geq 1$ , for $1 \leq i \geq n$

The distance between these two hypeplanes can be calculated as $\frac{2}{||\vec{W}||}$  

When $||\vec{W}|| == 1$, Then $\frac{2}{||\vec{W}||} == 2$  
When $||\vec{W}|| == 2$, Then $\frac{2}{||\vec{W}||} == 1$  
When $||\vec{W}|| == 4$, Then $\frac{2}{||\vec{W}||} == 0.5$  
here we can see that bigger the norm $||\vec{W}||$ smaller the distance between hyperplanes.  And smaller the norm $||\vec{W}||$ bigger the distance between hyperplanes.  

**Maximizing the margin is the same thing as minimizing the norm of w**. The goal is to maximize the margin. Among all possible hyperplanes meeting the constraints,  SVM will choose the hyperplane with the smallest $∥\vec{W}∥$ because it is the one which will have the biggest margin. This gives the **Optimization Problem** for SVM which is as follows,  

Minimize $||\vec{W}||$ such that **$y_i(w.x_i + b) \geq 1$** for any $i=1,2,3,....,n$  it can be written in quadratic form as 

**$Min \frac{1}{2}||\vec{W}||^2$** such that **$y_i(w.x_i + b) \geq 1$** for any $i=1,2,3,....,n$


An ideal SVM analysis should produce a hyperplane that completely separates points into two classes. However, perfect separation may not be possible, or it may result in a model with so many cases that the model does not classify correctly. In this situation SVM finds the hyperplane that maximizes the margin and minimizes the missclassifications. 

When a point is missclassified or it is on the wrong side of the margin then a penalty is added denoted by slack variable($\epsilon$)  
<img src='SVM_7.png' style="height: 300px">  

So the Optimization Problem can bewritten as ,  

**$Min \frac{1}{2}||\vec{W}||^2$ + $C \sum_{i=0}^n \epsilon_i$**   where $C$ is the tradeoff between margin width and missclasifications.  For larger values of C, a smaller margin will be accepted if the decision function is better at classifying all training points correctly. A lower C will encourage a larger margin, therefore a simpler decision function, at the cost of training accuracy.  


####  Soft Margin  

SVM is extended when data are not linearly classifiable. 

>**Linear Classifier** : When datap oints can be classified by a classifier which is linear combination input features ($Y = M.X + C$ is as linear classifier equation).  

>**Non Linear Classifier** : When datap oints can be classified by a classifier which is not linear combination input features.  

>**Kernel Function** : Is a function which takes input from finite dimensional space and transform them into an output in much higher dimensional space.
