### Tensorflow Fundamentals

##### What is a Tensor?

- When we represent data for machine learning, this generally needs to be done numerically. Especially when referring specifically of neural network data representation, this is accomplished via a data repository known as the tensor. A tensor is a container which can house data in N dimensions. 

- Tensors can hold data in any dimensions such as 0-dimension, 1-dimension (Vector), 2-dimensions(Matrix), 3-dimensions(Tensor) upto N-dimensions(Tensor)

##### Import Tensorflow

```python
import tensorflow as tf

print('Tensorflow Version ', tf.__version__)
```

###### Creating a Tensor

```python
# Using tf.constant

scalar = tf.constant(5)
scalar
print(scalar, ' \n Number of Dimensions of Scalar ', scalar.ndim)
```

```python
vector = tf.constant([8,6])
vector
print(vector, ' \n Number of Dimensions of Vector ', vector.ndim)
```

```python
matrix = tf.constant([[8,6],
                      [1,9]])
print(matrix, ' \n Number of Dimensions of Matrix ', matrix.ndim)
```

```python
tensor = tf.constant([[[8,6,1],
                      [9,9,1]],
                      
                      [[1,4,8],
                      [4,3,2]],
                      
                      [[5,8,6],
                      [7,9,2]]])
print(tensor, ' \n Number of Dimensions of Tensor ', tensor.ndim)
```

```python
# Using tf.Variable

tensor = tf.Variable([10,7])
tensor
```

```python
# Lets understand the difgference between tf.constant and tf.Variable
immutable_tensor = tf.constant([[1,8,9],
                                [6,3,5]])

mutable_tensor = tf.Variable([[10,5,12],
                              [17,13,19]])

immutable_tensor, mutable_tensor
```

```python
# Lets try to change an element of mutable_tensor

mutable_tensor[[0,2]].assign(2)

mutable_tensor
```

```python
# now lets try to change an element of immutable_tensor

immutable_tensor[[0,2]].assign(15)
```

**Note:** Elements of a tensor created using `tf.constant` can not changed after creation. Where as in case of a tensor created using `tf.Variable` x, its elements can be assigned to new value.

```python
# creating tensor with random set of values

# random tensor ( data forms  uniform distribution)
uniform_random_tensor = tf.random.uniform(shape=(3,3), dtype=tf.float16, seed=42)
print('Uniform Random Tensor: ', uniform_random_tensor )

# # random tensor ( data forms normal distribution)
normal_random_tensor = tf.random.normal(shape=(3,2,3), dtype=tf.float16, seed=42)
print('\nNormal Random Tensor: ', normal_random_tensor )
```

```python
# seed
tf.random.set_seed(5) # Global Seed: ensures reproduceability on every run
random_tensor_1 = tf.random.normal(shape=(2,2,2), dtype=tf.float16)
random_tensor_2 = tf.random.normal(shape=(2,2,2), dtype=tf.float16)

print(random_tensor_1, '\n\n', random_tensor_2)

# if we run the above code we will see same elements inside these tensors on every run
```

**Note:**

Random tensors ar tensors of arbitary size which holds random numbers generated from a distribution such as uniform distribution, normal distribution etc.

A neural network on its first training execution generate random weights. later on during the training it keeps on update these weight. this random weights are one of the use of the random tensors.

global seed can be set using `tf.random.set_seed()` to ensure same random tensors are generated on every run.
